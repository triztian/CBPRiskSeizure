After we inspect the brand list, we'll look for "high end" brands and then randomly select

We'll define a few helper functions to perform sampling and product selection for our company:

````python
HIGH_END_BRANDS = {"Nike", "H&M", "Calvin Klein Jeans", "Calvin Klein Underwear", "ADIDAS"}
SAMPLE_ITERATIONS = 5_000

def select_high_end_products(df, n=40):
    """Selects a number of products from high end brands"""
    if n <= 0 or n >= len(df):
        return df.copy()
    hep_df = df[df['brand_name'].str.contains("|".join(HIGH_END_BRANDS))].head(n=int(n))
    
    for _ in range(SAMPLE_ITERATIONS):
        results_df = hep_df.sample(n=int(n))

    return results_df
````

````python
def select_regular_products(df, n=40):
    """Select a sample of random regular products (non-high end)"""
    if n <= 0 or n >= len(df):
        return df.copy()

    return df[~df['brand_name'].str.contains("|".join(HIGH_END_BRANDS))][:int(n)].copy()
````

````python
def select_random_products(df, n=40):
    """Selects a number of random products"""
    if n <= 0 or n >= len(df):
        return df.copy()

    for _ in range(10_000):
        s = df.sample(n=int(n))

    return s.copy()
````

````python
# Define the distribution of item types
TOTAL_PRODUCTS = 100
high_end_pct, regular_pct, random_pct = (.2, .4, .4)

df = myntra_df.copy()

high_end_df = select_high_end_products(myntra_df, n=high_end_pct * TOTAL_PRODUCTS)

df = df[~df.product_id.isin(high_end_df.product_id)]
regular_df = select_regular_products(myntra_df, n=regular_pct * TOTAL_PRODUCTS)

df = df[~df.product_id.isin(pd.concat([high_end_df, regular_df]).product_id)]
random_df = select_random_products(myntra_df, n=random_pct * TOTAL_PRODUCTS)

selections_df = pd.concat(
    [high_end_df, regular_df, random_df]
)

selections_df.head()
````

````python
products_df = myntra_df[myntra_df.product_id.isin(selections_df.product_id)].copy()
products_df.head()
````

Next, we'll convert the prices into us dollar for consistency, we'll use the Federal Reserve's currency rate:

  * https://www.federalreserve.gov/releases/h10/current/

For July 26 the dollar to rupee rate is:

```
83.7000
```

````python
usd_to_rs = 83.7
````

Now, we'll add a new column that has the price in USD:

````python
products_df['price_usd'] = products_df['price_rs'].apply(lambda x: x / usd_to_rs)
````
